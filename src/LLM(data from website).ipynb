```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

# Model identifier
model_name = "meta-llama/Llama-2-7b-chat-hf"


token = True

tokenizer = AutoTokenizer.from_pretrained(model_name, token=token)

# Conditional parameters based on GPU availability
if torch.cuda.is_available():
    model_params = {
        "device_map": "auto",
        "torch_dtype": torch.float16,
        "token": token,
        "load_in_8bit": True,  # Enable 8bit loading only if GPU is available
        # "load_in_4bit": True,
    }
else:
    # Parameters for running without GPU
    model_params = {
        "device_map": "auto",
        "torch_dtype": torch.float16,
        "token": token,
    }


try:
    model = AutoModelForCausalLM.from_pretrained(model_name, **model_params)
except Exception as e:
    print(f"Failed to load the model: {e}")


def get_chatbot_response(question):
    # Tokenize the input question
    inputs = tokenizer(question, return_tensors="pt")
    # Generate a response using the model
    response_ids = model.generate(inputs["input_ids"], max_length=50, num_return_sequences=1)
    # Decode the response
    response = tokenizer.decode(response_ids[0], skip_special_tokens=True)
    return response


```

```python
pipe = pipeline("text-generation",
                model=model,
                tokenizer= tokenizer,
                torch_dtype=torch.bfloat16,
                device_map="auto",
                max_new_tokens = 512,
                do_sample=True,
                top_k=30,
                num_return_sequences=1,
                eos_token_id=tokenizer.eos_token_id
                )
```

```python
llm=HuggingFacePipeline(pipeline=pipe, model_kwargs={'temperature':0})
```

```python
llm.predict("Artificial Intelligence in Anesthesiology: Current Techniques, Clinical Applications, and Limitations")
```

#**Step 08: Initialize the Retrieval QA with Source Chain**

```python
chain = RetrievalQAWithSourcesChain.from_llm(llm=llm, retriever=vectorstore.as_retriever())
```
